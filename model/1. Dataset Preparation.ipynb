```python
depressive = pd.read_csv("../data/Depressive_Tweets.csv", encoding="latin-1")
random = pd.read_csv("../data/Random_Tweets.csv", encoding="latin-1")
validation = pd.read_csv("../data/Validation_Tweets.csv", encoding="latin-1")
```
